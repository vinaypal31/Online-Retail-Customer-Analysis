```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"] , format="%d/%m/%y %H:%M")

df["InvoiceDate"]
```

```python
df = df.set_index("InvoiceDate")
df.resample("ME")["InvoiceNo"].nunique()
```

```python
df.resample("ME")["CustomerID"].nunique().plot(figsize=(15,5),marker="s",ms="15",mfc="orange")
```

```python
df.head()
```

```python
newdf = df.groupby(["InvoiceNo" , "InvoiceDate"])["CustomerID"].first().reset_index()

newdf
```

```python
def custom_count(grp):
    a = grp.value_counts()
    data = a > 1
    return data.sum()

newdf.set_index("InvoiceDate",inplace=True)
monthly_repeate_customers = newdf["CustomerID"].resample("ME").apply(custom_count)
```

```python
df.plot(figsize=(15,5),label="unique customers")
monthly_repeate_customers.plot(figsize=(15,5),label="repeat customers")
plt.legend()
plt.show()
```
